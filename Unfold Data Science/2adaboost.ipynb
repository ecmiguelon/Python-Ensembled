```python
#Importing neccesary packages
# Load libraries
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

**1. Datos**

```python
df=pd.read_csv("income.csv")
df
```

target = income_level

**2. Separación de Datos**

```python
#target y features
X = df.iloc[:, 0:6] 
y = df.iloc[:,6]
```

```python
y.shape
```

```python
#Spliting data. Datos de entrenamiento y testeo 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 )
```

**3. Entrenamiento y Testeo**

```python
#objeto Adaboost
AdaModel = AdaBoostClassifier(n_estimators = 50, learning_rate=1, algorithm="SAMME")
```

```python
# Entrenamiento Adaboost Classifer
model = AdaModel.fit(X_train, y_train)

# Prediccción the response for test dataset
y_pred = model.predict(X_test)
```

base_estimator: It is a weak learner used to train the model. It uses DecisionTreeClassifier as default weak learner for training purpose. You can also specify different machine learning algorithms.

n_estimators: Number of weak learners to train iteratively.

learning_rate: It contributes to the weights of weak learners. It uses 1 as a default value.

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
